```python
# Task 1

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

y = np.sin(x)
z = np.cos(x)

plt.title("Графики синуса и косинуса")
plt.plot(x, y, label="y = sin x")
plt.plot(x, z, label="y = cos x")
plt.xlabel("Ось X", color="red", fontsize=14, labelpad=5)
plt.ylabel("Ось Y", color="red", fontsize=14, labelpad=3)


plt.legend()
plt.show()
```

```python
# Task 2

lst = [np.random.normal() for i in range(1000)]


plt.title("Гистограмма нормального распределения чисел до 1000")
plt.hist(lst, bins=20)
plt.xlabel("Ось X", color="red", fontsize=14, labelpad=5)
plt.ylabel("Ось Y", color="red", fontsize=14, labelpad=3)
plt.show()
```

```python
# Task 3

a = np.random.randint(1,10,size=(5,5))
print(a)
print(np.mean(a))
print(np.max(a))
print(np.min(a))
print(np.sum(a,axis=0)) # По столбцам
print(np.sum(a,axis=1)) # По строкам
```

```python
# Task 4

plt.imshow(a, cmap="autumn", interpolation ='nearest', origin ='upper')
plt.colorbar(label="Число", orientation="horizontal")
for i in range(5):
  for j in range(5):
    plt.text(i,j,a[j][i], horizontalalignment='center',
     verticalalignment='center')
plt.show()
```

```python
# Task 5

x1 = np.arange(-10,10,0.001)
y1 = x1 ** 2
plt.subplot(3,1,1)
plt.plot(x1,y1, color="#217043")

x2 = np.random.randint(-100,100,size=20)
y2 = np.random.randint(-100,100,size=20)
plt.subplot(3,1,2)
plt.scatter(x2, y2, color="#be8fdb")

x3 = [f"A{i}" for i in range(1, 11)]
y3 = np.random.randint(1, 10, len(x3))
plt.subplot(3,1,3)
plt.bar(x3, y3, color="#a87732")

plt.subplots_adjust(top=5, bottom=2, hspace=0.1)
plt.show()
```
